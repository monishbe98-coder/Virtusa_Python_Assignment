```python
Menu Driven Program
```

```python
while True:
    print("\n1. CM to Feet")
    print("2. KM to Miles")
    print("3. USD to INR")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        cm = float(input("Enter length in CM: "))
        feet = cm / 30.48
        print("Feet =", round(feet, 2))

    elif choice == 2:
        km = float(input("Enter distance in KM: "))
        miles = km * 0.621371
        print("Miles =", round(miles, 2))

    elif choice == 3:
        usd = float(input("Enter USD amount: "))
        rate = float(input("Enter current USD to INR rate: "))
        inr = usd * rate
        print("INR =", round(inr, 2))

    elif choice == 4:
        print("Exiting...")
        break

    else:
        print("Invalid Choice")
```

```python
Reverse Number Pattern
```

```python
n = int(input("Enter number: "))

for i in range(n, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```

```python
Subtract 7 Days from Given Date
```

```python
from datetime import datetime, timedelta

date_str = input("Enter date (YYYY-MM-DD): ")

given_date = datetime.strptime(date_str, "%Y-%m-%d")

new_date = given_date - timedelta(days=7)

print("Date after subtracting 7 days:")
print(new_date.date())
```

```python
Add 7 Days and 12 Hours
```

```python
from datetime import datetime, timedelta

date_str = input("Enter datetime (YYYY-MM-DD HH:MM:SS): ")

given_date = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")

new_date = given_date + timedelta(days=7, hours=12)

print("Updated DateTime:")
print(new_date)
```

```python
Calculate Date 4 Months from Given Date
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

date_str = input("Enter date (YYYY-MM-DD): ")

given_date = datetime.strptime(date_str, "%Y-%m-%d")

future_date = given_date + relativedelta(months=4)

print("Date after 4 months:")
print(future_date.date())
```

```python
Count Frequency of Each Character (Ignore Spaces, Case-Insensitive)
```

```python
def count_characters(text):
    counts = {}

    text = text.lower().replace(" ", "")

    for ch in text:
        counts[ch] = counts.get(ch, 0) + 1

    return counts

text = input("Enter a string: ")

result = count_characters(text)

print(result)
```

```python
Check Whether Two Strings Are Anagrams
```

```python
def is_anagram(str1, str2):
    str1 = str1.lower().replace(" ", "")
    str2 = str2.lower().replace(" ", "")

    return sorted(str1) == sorted(str2)

word1 = input("Enter first word: ")
word2 = input("Enter second word: ")

print(is_anagram(word1, word2))
```

```python
Extract Strings Longer Than 5 Characters and Starting with a Vowel
```

```python
words = input("Enter words separated by space: ").split()

result = [
    word
    for word in words
    if len(word) > 5 and word[0].lower() in "aeiou"
]

print(result)
```

```python
Invert Dictionary (Author → Books to Book → Author)
```

```python
authors = {}

n = int(input("Enter number of authors: "))

for _ in range(n):
    author = input("Enter author name: ")

    books = input(
        "Enter books separated by comma: "
    ).split(",")

    authors[author] = [book.strip() for book in books]

inverted = {}

for author, books in authors.items():
    for book in books:
        inverted[book] = author

print(inverted)
```
