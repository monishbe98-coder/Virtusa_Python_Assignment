```python
reverse 4 digit number , check reverse is true
```

```python
num = int(input("Enter a 4-digit number: "))

original = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number:", rev)
print(original == rev)
```

```python
Armstrong number check
```

```python
num = int(input("Enter a number: "))

original = num
n = len(str(num))
armstrong_sum = 0

while num > 0:
    digit = num % 10
    armstrong_sum += digit ** n
    num //= 10

if armstrong_sum == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
Hours , minutes , Seconds in number of days
```

```python
days = int(input("Enter number of days: "))

hours = days * 24
minutes = days * 24 * 60
seconds = days * 24 * 60 * 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

```python
simple interest
```

```python
principal = float(input("Enter principal amount: "))
roi = float(input("Enter rate of interest (%): "))
time = float(input("Enter time (years): "))

simple_interest = (principal * roi * time) / 100

print("Simple Interest = {:.2f}".format(simple_interest))
```
