```python
odd and even lists
```

```python
list1 = [10, 21, 33, 44, 55]
list2 = [11, 22, 34, 45, 60]

new_list = []

for num in list1:
    if num % 2 != 0:
        new_list.append(num)

for num in list2:
    if num % 2 == 0:
        new_list.append(num)

print("New List:", new_list)
```

```python
Find HCF (GCD) of two numbers
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

small = min(num1, num2)

for i in range(small, 0, -1):
    if num1 % i == 0 and num2 % i == 0:
        hcf = i
        break

print("HCF =", hcf)
```

```python
Calculate In-Hand Salary
```

```python
salary = float(input("Enter annual salary: "))

hra = salary * 0.10
da = salary * 0.05
pf = salary * 0.03

if salary <= 100000:
    tax = 0
elif salary <= 1000000:
    tax = salary * 0.10
elif salary <= 2000000:
    tax = salary * 0.20
else:
    tax = salary * 0.30

in_hand_salary = salary - (hra + da + pf + tax)

print("HRA =", hra)
print("DA =", da)
print("PF =", pf)
print("Tax =", tax)
print("In-Hand Salary =", in_hand_salary)
```

```python
Count Letters, Digits and Special Characters
```

```python
str1 = input("Enter a string: ")

chars = 0
digits = 0
symbols = 0

for ch in str1:
    if ch.isalpha():
        chars += 1
    elif ch.isdigit():
        digits += 1
    else:
        symbols += 1

print("Chars =", chars)
print("Digits =", digits)
print("Symbols =", symbols)
```

```python
Find the last position of a substring
```

```python
str1 = "Milind is a data scientist who knows Python. Milind works at google."
substr = "Milind"

index = str1.rfind(substr)

if index != -1:
    print("Last occurrence starts at index", index)
else:
    print("Substring not found")
```

```python
Remove all characters except integers
```

```python
str1 = "I am 25 years and 10 months old"

result = ""

for ch in str1:
    if ch.isdigit():
        result += ch

print(result)
```

```python
Find words with both alphabets and numbers
```

```python
text = input("Enter a string: ")

words = text.split()

for word in words:
    has_alpha = False
    has_digit = False

    for ch in word:
        if ch.isalpha():
            has_alpha = True
        elif ch.isdigit():
            has_digit = True

    if has_alpha and has_digit:
        print(word)
```

```python
Replace each special symbol with '#'
```

```python
str1 = "/*Jon is @developer & musician!!"

result = ""

for ch in str1:
    if ch.isalnum() or ch == " ":
        result += ch
    else:
        result += "#"

print(result)
```

```python
Slice list into 3 equal chunks and reverse each chunk
```

```python
sample_list = [11,49,8,23,14,12,78,45,89]

size = len(sample_list) // 3

for i in range(0, len(sample_list), size):
    chunk = sample_list[i:i+size]
    print(chunk[::-1])
```

```python
Count occurrence of each element in a list
```

```python
sample_list = [11, 45, 8, 11, 23, 45, 23, 45, 89]

count_dict = {}

for item in sample_list:
    if item in count_dict:
        count_dict[item] += 1
    else:
        count_dict[item] = 1

print(count_dict)
```
